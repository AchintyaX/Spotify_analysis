```python
# reading the dataset 

import pandas as pd 

spotify = pd.read_csv('SpotifyFeatures.csv')
df = spotify
df
```

### Making all the data numerical 

```python
modes = {
    'Major':1,
    'Minor':0
}

df['mode'] = [modes[i] for i in df['mode']]
df.head()
```

```python
print(df.key.unique())
print(len(df.key.unique()))
```

```python
keys = {
    'A':0,
    'B':1,
    'C':2,
    'D':3,
    'E':4,
    'F':5,
    'G':6,
    'A#':7,
    'C#':8,
    'D#':9,
    'F#':10,
    'G#':11
}
df.key =[keys[i] for i in df.key]
```

```python
genres = {
    'Movie':0,
    'R&B':1,
    'A Capella':2,
    'Alternative':3,
    'Country':4,
    'Dance':5,
    'Electronic':6,
    'Anime':7,
    'Folk':8,
    'Blues':9,
    'Opera':10,
    'Hip-Hop':11,
    "Children's Music":12,
    'Rap':13,
    'Indie':14,
    'Classical':15,
    'Pop':16,
    'Reggae':17,
    'Reggaeton':18,
    'Jazz':19,
    'Ska':20,
    'Comedy':21,
    'Soul':22,
    'Soundtrack':23,
    'Rock':24,
    'World':25,
    'Children’s Music':12
}
df.genre = [genres[i] for i in df.genre]
df
```
